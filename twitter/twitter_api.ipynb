```python
import urllib.parse
import oauth2 as oauth
import json
import config
```

### Functions

```python
def oauth_req(url, key, secret, http_method="GET", post_body=b"", http_headers=None):
    consumer = oauth.Consumer(key=config.CONSUMER_KEY, secret=config.CONSUMER_SECRET)
    token = oauth.Token(key=key, secret=secret)
    client = oauth.Client(consumer, token)
    resp, content = client.request( url, method=http_method, body=post_body, headers=http_headers )
    content_json = json.dumps(json.loads(content), indent=4, sort_keys=True)
    return content_json
```
