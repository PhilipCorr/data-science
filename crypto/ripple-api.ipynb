```python
import pandas as pd
import requests
import json
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_style("darkgrid")

```

```python
stats_payload = {'start':'2017-01-01T00:00:00Z','limit':1000}
```

```python
stats_response = requests.get('https://data.ripple.com/v2/stats', params=stats_payload)
```

```python
stats_response.json()['count']
```

```python
stats = stats_response.json()['stats']
```

```python
stats[0:10]
```

```python
dates = []
payments_count = []
for stat in stats:
    dates.append(stat['date'])
    payments_count.append(stat['metric']['payments_count'])
    
```

```python
dates
```

```python
plt.plot(dates, payments_count)
plt.show()
```
